```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import copy
```

```python
path = './sample/3.jpg'
img = cv2.imread(path)
img = cv2.GaussianBlur(img,(5,5),0)
r = int(np.sqrt(np.size(img))/100)
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV) 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#default setting for red mask
#lower_red=np.array([169, 100, 100])             
#upper_red=np.array([189, 255, 255])

lower_red=np.array([169, 100, 100])             
upper_red=np.array([189, 255, 255])

mask_red=cv2.inRange(hsv,lower_red,upper_red)
res_red=cv2.bitwise_and(img,img,mask=mask_red)
```

```python
plt.imshow(res_red)
```

```python
plt.imshow(mask_red)
```

```python
kernel = np.ones((10,10), np.uint8)
img_dilation = cv2.dilate(mask_red, kernel, iterations=1)
plt.imshow(img_dilation)
```

```python
img_erosion = cv2.erode(img_dilation, kernel, iterations=1)
plt.imshow(img_erosion)
```

```python
imgray = img_erosion
ret,thresh = cv2.threshold(imgray,10,255,0)
im2, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
contours_ = []

#reject small contour areas
for c in contours:
    if cv2.contourArea(c) >= 500:
        contours_.append(c)
length = len(contours_)
cv2.drawContours(img, contours_, -1, (0,255,0), 3)
plt.imshow(img)
```

```python
#points = np.argwhere(mask_red)
points = np.array(contours_)
center = []
if length:
    for i in range(0, length):
        x = int(np.mean(points[i][:,0,0]))
        y = int(np.mean(points[i][:,0,1]))
        cv2.circle(img, (x, y),r,(0,255,0),-1)
        center.append([x,y])
plt.imshow(img)
```

```python
x = (np.array(center))[:,0]
y = (np.array(center))[:,1]
coefficients = np.polyfit(x,y,2)

a = np.arange(img.shape[1])
f = np.poly1d(coefficients)
result = np.zeros((np.size(a),2))
result[0:,0] = a
result[0:,1] = f(a)
result = result.astype(int)
result = result[np.where(result[:,1]>=0)]
result = result[np.where(result[:,1]<img.shape[0])]

img = cv2.polylines(img, [result], True, (255,255,255),int(r/2))

plt.imshow(img)
```
